# 자료형


1. 데이터 저장방법에 따른 자료형의 분류  
  직접(direct)형 : 직접 데이터를 표현하는 자료형 : int, float, complex      
  시퀀스(sequence)형: 다른 데이터를 포함하는 자료형, 순서가 있는 집합 자료형 : list,str, tuple, bytes, bytearray,range    
  매핑(Mapping)형 : 다른 데이터를 포함하는 자료형 : dict    
  집합(Set)형 : 순서가 없고, 중복된 값도 없다. : set, frozenset  


2. 변경 가능성에 따른 자료형 분류  
  변경 가능형(Mutable) : 데이터의 값을 변경할 수 있다.  : list, dict, set  
  변경 불가능형(Immutable) : 데이터의 값을 변경할 수 없다. : int, float, complex, str, tuple, frozenset


3. 데이터 저장 개수에 따른 자료형의 분류  
  리터널(Literal/Scalar)형 : 한 가지 객체만 저장한다. : int, bytes, str, float, complex  
  컨테이너(Container)형 : 여러 가지 객체를 저장한다. : list, tuple, dict, set, fronzenset

문자열포매팅(서식지정) 형식


|코드|설명 |
|---|:---:|
|%d|정수|
|%f|실수|
|%s|문자열|
|%c|하나의문자|
|%o|8진수|
|%x|16진수|
|%%|%기호자체를출력하고자할때|

```python
s_val7 = 'python'
s_val8 = 83.5


print('%s 의 점수는 %.1f입니다.' % ( s_val7, s_val8)) # python1, 2방식
print('{} 의 점수는 {}입니다.'.format(s_val7, s_val8)) # python3 방식
print('{1} 의 점수는 {0}입니다.'.format(s_val8, s_val7))
print('{subject} 의 점수는 {jumsu}입니다.'.format(jumsu=s_val8, subject=s_val7))
print('{{0}} 의 점수는 {{1}}입니다.'.format(s_val7, s_val8)) # { }자체를 출력할때 {{}}을 한다.
print('{{{0}}} 의 점수는 {{{1}}}입니다.'.format(s_val7, s_val8)) # {}자체와 값을 출력할때는 {{{}}}을 한다.


print('{0:<10} 의 점수는 {1}입니다.'.format(s_val7, s_val8))  # 10자리 확보하고 남은 공간은 오른쪽을 여백으로 채운다.
#print('{:<10} 의 점수는 {}입니다.'.format(s_val7, s_val8))


print('{0:>10} 의 점수는 {1}입니다.'.format(s_val7, s_val8)) 
```

list는 한 변수에서 데이터를 각각 꺼내는 언팩킹(unpacking) 지원을 한다.

```python
fruits = ['apple', 'mango', 'grape', 'banana']
print(fruits)

zero, one, two, three = fruits
print(zero, one, two, three)

zero, _, one, _ = fruits
print(zero, one)

zero, *group = fruits # *group을 사용해서 뒤에 값들을 한번에 가져옴 -> 리스트
print(zero,group)
print(type(zero),type(group)) # <class 'str'> <class 'list'>

# 1번만 사용 가능
# zero, *group1, *group2 = fruits
# print(zero,group1, group2)

*group, zero = fruits
print(group,zero) # ['apple', 'mango', 'grape'] banana

zero, *group, three = fruits
print(zero, group, three) # apple ['mango', 'grape'] banana
```

## 집합(set)
- 멤버쉽(원소와 집합사이의 귀속관계)검사, 시퀀스에서의 중복제거와 교집합,차집합,합집합 등 수학연산 수행에 사용된다
- 순서가 유지 안되므로 인덱싱과 슬라이싱 사용을 사용할 수 없다
- 반복 가능(iterable)한 객체로 집합 객체를 생성할 수 있다.

```python
s = set()
print(s,type(s))
print(dir(s))
```

```python
# 순서 유지가 안되고 중복값 허용 안되며 오름차순으로 정렬이 된다
set((3,1,5,2,5))
```

```python
a = (3,1,5,2,5)
b = set(a)

print(b)
```

```python
# set에 문자열 값도 넣을 수 있다
set('apple')
```

```python
# 리스트도 가능
set([3,1,5,2,5])
```

```python
a = [3,1,5,2,5]
b = set(a) # a값을 복사해서 b에 저장 (deep copy) 서로 영향이 없음
print(b)

a[0] = 6
print(a)
print(b)
```

```python
k = [2,4]
a = [3,1,k,5]
print(a)

# a 안에 k라는 변경 가능한 데이터가 있기 때문에 생성 불가능
b = set(a)
print(b)
```

```python
# dict 자료형
# key : value 한쌍
# dict 자료형으로 set을 생성하면 키 값만 참조된다
b = set({'one':1, 'two':2})
print(b,type(b))
```

```python
a = set((1,2,3,4,5))
b = set((4,5,6,7))

print(a.union(b)) # 합집합 메소드
print(a | b) # 합집합 기호

print(a.intersection(b)) # 교집합 메소드
print(a & b) # 교집합 기호

print(a.difference(b)) # 차집합 메소드
print(a-b) # 차집합 기호
```

```python
a = set((1,2,3))
a.add(4) # add - 한개만 추가
print(a)
```

```python
a = set((1,2,3))
a.update((4,5)) # update - 여러개 추가 (iterable)
print()
```

```python
a = set((1,2,3))
a.remove(2) # () 괄호 안에 요소를 제거
print(a)
```

- frozenset은 frozenset()으로 생성한다

## dict
- 특정 키를 주면 이와 관련된 값을 돌려주는 내용 기반으로 검색하는 자료형이다.
- 데이터의 저장 순서가 없다.
- 내부적으로 해시 기법을 사용하기 때문에 검색 속도가 빠르다.
- key-value형식으로 저장을 하기 때문에 key을 이용해서 value에 접근한다.
- key 값은 중복을 허용하지 않는다.

#### zip()함수를 이용해서 dict()생성할 수 있다

```python
keys=['one', 'two', 'three']
values=[1,2,3]

a = zip(keys, values) # zip()로 묶기
print(a)
d = dict(a) # dict 생성
print(d)
print(d['one'])
```

```python
keys=['one', 'two', 'three','four','five']
values=[1,2,3]

a = zip(keys, values) # keys, values의 크기가 다르면 작은수에 맞춰 생성된다
print(a)
d = dict(a)
print(d)
print(d['one'])
```

```python
d = {'one':1, 'two':2}
print(d.keys()) # dict에서 key값만 가져오기
print(d.values()) # dict에서 value값만 가져오기
print(d.items()) # dict에서 key와 value 가져오기 튜플로 묶어서 한쌍 만들고 리스트로 반환

# get은 복사해서 가져옴
print(d.get('one')) # 해당 키의 value 값 가져오기
print(d.get('three')) # 해당 키가 없다면 none 반환
print(d.get('three',0)) # 해당 키가 없다면 직접 설정한 값 반환도 가능

d = {'one':1, 'two':2}
d.pop('two') # 메모리에서 아예 제거 , 제거되는 값 반환 가능
print(d)

d = {'one':1, 'two':2}
del d['one'] # 메모리에서 아예 제거 , 제거되는 값이 반환 X
print(d)

d.clear()
print(d)
```
