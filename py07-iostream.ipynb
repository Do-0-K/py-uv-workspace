## 파일 입출력


- 파이썬 표준 함수(내장함수)에는 파일과 관련된 처리를 하는 함수가 기본으로 제공된다.
- 파일은 텍스트 파일과 바이너리 파일로 나뉜다.
- 파일을 열때 open( )함수를 사용한다.
- 파일 입출력 순서
 ① 어플리케이션이 운영체제에게 파일 처리를 API 함수를 통해 의뢰하면, 운영체제가
   요청한 업무를 수행해 주고 그 결과를 어플리케이션에게 돌려줌.
 ② 어플리케이션이 파일 처리 업무를 의뢰하는 과정과 각 과정에서 사용되는 파이썬
    함수

| 단계 | 처리 흐름 | 매칭 소스 코드 | 세부 설명 |
|---|---|---|---|
| ① | 파일 열기 | file = open() | open() 내장함수로 파일을 연다. |
| ② | 파일 읽기/쓰기 | file.read() / file.write() | 얻어진 파일 객체에서 데이터를 읽고 쓴다.(read,write) |
| ③ | 파일 닫기 | file.close() | close() 함수로 객체의 사용을 종료한다. |

파일객체 = open('파일경로', '파일열기모드')  

| 파일 열기 모드 | 기능 |
|---|---|
|w:쓰기모드|파일에 내용을 쓸|
|a:추가모드 | 파일에 내용을 추가할때 사용 |

파일 읽기/쓰기 함수

| 구분 | 함수(메서드)명 | 반환 및 특징 |
| :---: | :--- | :--- |
| **읽기 (Read)** | `read()` | 파일 전체 내용을 하나의 문자열로 반환 |
| **읽기 (Read)** | `readline()` | 파일의 현재 위치에서 딱 한 줄만 반환 |
| **읽기 (Read)** | `readlines()` | 모든 줄을 읽어 문자열 리스트로 반환 |
| **쓰기 (Write)** | `write()` | 문자열 하나를 파일에 기록 |
| **쓰기 (Write)** | `writelines()` | 문자열 리스트/튜플을 한 번에 기록 |

````python
fread = open('source/sample.txt','r') # 파일을 읽기로 오픈

while True:
    line = fread.readline() # readline()은 파일에서 한 줄씩 읽어옴
    if not line: # 파일의 내용을 다 가져왔다면
        break    # while문을 빠져나가라
    print(line, end = '')

fread.close()
````

````python
fread = open('source/sample.txt', 'r')
line = fread.readlines() # readlines()는 파일에서 한 라인씩 가져와서 list에 저장한 후 return
print(line)
fread.close()

for data in line:
    print(data,end='')
````

### with문과 함께 사용
- with문과 함께 파일을 사용하면 읽기, 쓰기할 때 close()를 안해도 된다.

````python
with open('source/sample.txt', 'a') as f: # a는 append 추가라는 의미 (업데이트 X, 추가 O)
    for i in range(6,11):                   # 파일에 새로운 내용을 추가한다.
        data = f'{i}번째 줄입니다.\n'
        f.write(data)
````

## 주요 입출력(직렬화/데이터 저장) 모듈
직렬화(Serialization)란?  
- 파이썬의 객체(리스트, 딕셔너리, 클래스 등)를 파일로 저장하거나     네트워크로 전송할 수 있는 바이트/텍스트 형태로 변환하는 작업입니다.
- 입출력 모듈 종류  
①pickle: 텍스트 상태의 데이터가 아닌 파이썬 객체 자체를 파일로 저장하는 모듈  
②json : 타 언어 호환성이 가장 뛰어난 텍스트 기반 모듈    
③csv : 엑셀과 연동되는 표 형태 데이터 전용 모듈  
④marshal : 파이썬 내부 바이트코드를 위한 고속 모듈    
⑤shelve : 파일 자체를 딕셔너리처럼 쓰는 영구 저장소 모듈  
⑥configparser : 설정 파일(INI)을 관리하는 전용 모듈    

| 모듈명 | 데이터 형태 | 타 언어 호환 | 보안성 | 주요 용도 |
| :---: | :---: | :---: | :---: | :--- |
| **pickle** | 바이너리 | 불가능 (파이썬 전용) | ⚠️ 취약 (경고) | 파이썬 복잡한 객체 완벽 저장 |
| **json** | 텍스트 (문자열) | **우수 (글로벌 표준)** | 안전함 | 웹 API 통신, 일반 데이터 저장 |
| **shelve** | 바이너리 | 불가능 (파이썬 전용) | ⚠️ 취약 (경고) | 대용량 key-value 로컬 데이터베이스 |
| **csv** | 텍스트 (문자열) | **우수 (엑셀 연동)** | 안전함 | 표 형태 데이터, 데이터베이스 내보내기 |

### pickle 모듈이란?  
- 파이썬 객체 구조를 그대로 바이트 스트림으로 변환(직렬화)하고, 이를 다시 파이썬 객체로 복원(역직렬화)하는 파이썬 고유의 표준 모듈

- 핵심 함수  
 ①pickle.dump(obj, file): 객체를 바이너리 파일에 저장  
 ②pickle.load(file): 바이너리 파일로 부터 객체를 복원

- ⚠치명적인 보안 경고 (중요)  
①pickle은 신뢰할 수 없는 데이터(인터넷에서 다운로드한 파일 등)를 역직렬화할 때    악성 코드가 자동으로 실행될 수 있는 심각한 보안 취약점이 있다.  
②따라서 외부 통신용 데이터 저장에는 절대로 사용을 지양해야 한다.

````python
import pickle
````

````python
with open('source/sample.txt','wb') as f:
    pickle.dump(['apple','banana','lemon'],f) #dump(저장할 데이터, 파일)
````

````python
with open('source/sample.txt','rb') as f:
    print(pickle.load(f))
````

````python
with open('source/sample.txt','wb') as f:
    pickle.dump(['apple',20 ,True ],f) #dump(저장할 데이터, 파일)
````

````python
with open('source/sample.txt','rb') as f:
    print(pickle.load(f))
````

````python
with open('source/sample.txt','wb') as f:
    pickle.dump(('apple', 20, True),f) #dump(저장할 데이터, 파일)
````

````python
with open('source/sample.txt','rb') as f:
    print(pickle.load(f))
````

###  json (웹 표준 및 안전한 대체재)
- json 모듈?

  ① 텍스트 기반(문자열)으로 저장되어 사람이 직접 읽고 수정할 수 있음

  ②자바스크립트, 자바, C++ 등 다른 프로그래밍 언어와 데이터 교환 시 표준으로 사용
  
  ③텍스트 형식이므로 pickle과 같은 악성 코드 실행 위험이 없어 매우 안전함
- 핵심 암기 공식: 's' = String (문자열)
  - s가 없는 함수 (dump, load)   →  '파일(File)' 객체를 직접 읽고 쓸 때 사용
  - s가 있는 함수 (dumps, loads) →  '문자열(String)' 데이터를 직접 다룰 때 사용

- 4대 함수 기능 요약

  ① json.dump()  : 파이썬 객체를 'JSON 파일'로 저장 (입력)

  ② json.dumps() : 파이썬 객체를 'JSON 문자열'로 변환 (변환)

  ③ json.load()  : 'JSON 파일'을 파이썬 객체로 읽어옴 (출력)

  ④ json.loads() : 'JSON 문자열'을 파이썬 객체로 변환 (복원)

- json.dump() / dumps() 핵심 옵션
 - 파이썬 객체를 JSON 형식으로 변환할 때 가독성, 정렬, 한글 처리, 데이터 예외 등을
    세밀하게 제어하기 위해 인자(Argument) 값을 추가할 수 있다.

    ① indent : 들여쓰기 공백 수 지정 (가독성 향상)

    ② ensure_ascii : 아스키(ASCII) 변환 여부 (한글 깨짐 방지)

    ③ sort_keys : 딕셔너리 키(Key) 정렬 여부 (데이터 정렬)

    ④ separators : 데이터 구분 기호 변경 (용량 압축)

    ⑤ default : 직렬화 불가능한 객체의 처리 함수 지정 (에러 방지)

***문자열 전용 함수 (dumps / loads) 예제***
- 파일 입출력(I/O) 과정 없이, 메모리 상에서 파일 형식을 가진 문자열로만 주고받을 때 사용합니다. (웹 통신, API 전송 등)

````python
import json

data = {"name": "kim", "age": 25}

json_string = json.dumps(data)  # dict -> str
print(json_string)
print(type(json_string))  # str
````

````python
json_dict = json.loads(json_string)
print(json_dict)
print(type(json_dict))  # <class 'dict'>
````

````python
import json
from datetime import datetime

data = {
    "c_group": "인공지능팀",
    "a_id": 1004,
    "d_data": [1, 2, 3],
    "b_data": datetime.now(),
}

print(datetime.now())

# datatime 객체를 문자열로 변환해 주는 함수
def json_default_handler(obj):
    if isinstance(obj, datetime):
        return obj.strftime("%Y-%m-%d %H:%M:%S")

# 모든 옵션을 집약하여 호출
result = json.dumps(
    data,
    indent=4,  # 보기 좋게 설정하는 것
    ensure_ascii=False,  # 문자열을 아스키 코드로 표현 (True일때)
    sort_keys=True,  # 정렬
    default=json_default_handler,
)

print(result)
````

```
■ 콘솔 출력 내용 (Console Output)
  {
      "a_id": 1004,
      "b_date": "2026-07-15 15:14:49",
      "c_group": "인공지능팀"
  }
 
  ※ 특이사항 분석
  - 키가 알파벳 순서(a_id -> b_date -> c_group)로 자동 정렬되었습니다.
  - 한글 '인공지능팀'이 깨지지 않고 정상 표기됩니다.
  - 변환 불가였던 datetime 객체가 지정한 문자열 포맷으로 안전하게 출력되었습니다.
```

````python
import json

header = ["학번", "이름", "전공"]
students = [
    ["202601", "홍길동", "컴퓨터공학"],
    ["202602", "이순신", "데이터사이언스"],
    ["202603", "유관순", "인공지능학"],
]

dic = [{k: v for k, v in zip(header, student)} for student in students]

result = json.dumps(
    dic,
    indent=4,  # 보기 좋게 설정하는 것
    ensure_ascii=False,  # 문자열을 아스키 코드로 표현 (True일때)
)

print(result)


# 강사님 코드
data = [dict(zip(header,student)) for student in students]

print(data)
````

````python
with open('source/students.json','w') as f:
    json.dump(data, f,ensure_ascii=False, sort_keys=True)
````

````python
with open('source/students.json','r') as f:
    loaded_data = json.load(f)

print(loaded_data)
print(type(loaded_data))
````

### csv (Comma-Separated Values)
  - 표 형태의 데이터를 쉼표(,)로 구분하여 저장하는 표준 모듈
  - 엑셀(Excel)이나 데이터 분석 라이브러리(Pandas)와 완벽하게 연동됨
  - newline='' 옵션: 윈도우 환경 등에서 데이터 저장 시 줄바꿈이 중복되는 현상을 방지
  - encoding='utf-8-sig': 한글 데이터가 엑셀(Excel)에서 깨지지 않도록 인코딩 적용

````python
import csv

header = ["학번", "이름", "전공"]
students = [
    ["202601", "홍길동", "컴퓨터공학"],
    ["202602", "이순신", "데이터사이언스"],
    ["202603", "유관순", "인공지능학"],
]

# csv로 데이터 저장
# 저장을 할때 새로운 라인이 추가되지 않도록 newline=''을 지정
with open("source/records.csv", "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(header)  # 헤더작성 (제목)
    writer.writerows(students)  # 전치 데이터 작성

# csv 파일 내용 읽어오기
with open("source/records.csv","r",encoding='utf-8') as f:
    reader = csv.reader(f)
    for i in reader:
        print(f'학번:{i[0]} | 이름:{i[1]} | 전공:{i[2]}')
````
