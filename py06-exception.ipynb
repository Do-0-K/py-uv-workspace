## 예외처리
- 예외처리는 에러가 발생해되 실행을 중단하지 않고 계속 실행하게 해주는 방법이다.
- 예외처리 구조
```
 try:
    실행할 코드
 except:
    오류가 발생했을 때 처리하는 코드
 else:
    오류가 없을 때 처리하는 코드
 finally:
    무조건 처리하는 코드      
```
- 파이썬에서는 선언할 때 오류를 일일이 체크하지 않고, 사용할 때 에러가 발생하면 디버깅하는 스타일이다.
- EAFP
   - 허락보다는 용서를 구하기가 쉽다.(Easier to ask for forgiveness than permission)
   - 출처 : https://docs.python.org/3/glossary.html


- except 뒤에 에러 이름을 넣으면, 해당하는 에러가 발생했을 때만 except를 발생시킨다.
- Built-in Exceptions 예외 계층도 : https://docs.python.org/3/library/exceptions.html#exception-hierarchy

````python
a = 10 / 0
````

````python
# try 블록이 정상적으로 실행되지 않으면 else 블록은 실행되지 않는다
# try 영역에서 예외가 발생했을때 : try 실행 -> except 실행 -> finally 실행
# try 영역이 정상적으로 수행했을 때 : try 실행 -> else 실행 -> finally 실행
try:
    a = int(3)
    b = int('a')
    c = int(4)
except: # b에 문자가 들어왔으니 예외 발생
    print('숫자가 아닌 문자가 입력되었습니다')
else:
    print(f'{a}와 {b}와 {c}의 합 : {a+b+c}')
finally:
    print('program end')
````

## 다중 except
- 다중 except 선언이 가능하다.
- except뒤에 오는 as는 try 블록에서 발생한 이름을 받아온다.
```
  try:
     실행할 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드  
  except:
     예외가 발생했을 때 처리하는 코드  
```

````python
y = [10,20,30]

try:
    index, x = map(int, input('인덱스와 나눌 숫자를 입력하세요.').split())
    print(y[index] // x)
except ZeroDivisionError as e:
    print('숫자를 0으로 나눌 수 없습니다.', e)
except IndexError as e:
    print('잘못된 인덱스 입니다.', e)
except Exception as e:
    print(e)
else:
    print('정상처리 되었습니다.')
finally:
    print('program end')
````

````python
class Bird:
    def fly(self):
        # Bird를 상속받은 자식클래스들이 fly 함수를 오버라이딩 하도록 하는 작업
        # 추상 클래스로 만들어서 추상 함수로 데코레이터 하는 것과 같은 효과인 듯
        raise NotImplementedError

class Eagle(Bird):
    def display(self):
        print('display')
    
    def fly(self): # fly() 오버라이딩
        print('very fast')

eagle = Eagle()
eagle.fly()
````

````python
from abc import abstractmethod

class Bird:
    @abstractmethod
    def fly(self):
        pass

class Eagle(Bird):
    def display(self):
        print('display')
    #def fly(self):
    #    print('very fast')

eagle = Eagle()
eagle.fly()
````

````python
from abc import ABC, abstractmethod

class Bird(ABC):
    @abstractmethod
    def fly(self):
        pass

class Eagle(Bird):
    def display(self):
        print('display')
    #def fly(self):
    #    print('very fast')

eagle = Eagle() # ABC를 상속받은 Bird 추상클래스이다. Bird을 상속받은 Eagle도 추상클래스이다.
eagle.fly()
````

````python
class UserClass(Exception):
    def __str__(self):
        return '허용하지 않는 데이터 입니다'
````

````python
x = int(input('숫자 입력 : '))
print(x)
````

````python
try:
    x = int(input("숫자 입력 : "))
    if x <= 0:
        raise UserClass
    print(x)
except Exception as e:
    print(e)
except UserClass as e:
    print(e)
except ValueError as e:  # 다중 except를 사용할때 부모 클래스는 제일 아래에 위치
    print(e)
````

````python
def test():
    try:
        print("try 실행1")
        return
        print("try 실행2")

    except Exception as e:
        print("except", e)
    else:
        print("else")
    finally:
        print("finally")
````
