```
[문제1]
분자(a)와 분모(b)를 받아 나눗셈 연산을 하되, 분모가 0일 때 프로그램이 정지하지 않고 "0으로 나눌 수 없습니다."를 반환하는 proc_divide 함수를 구현하세요.

[출력결과]
5.0
0으로 나눌 수 없습니다.


```

````python
def proc_divide(n1, n2):
    try:
        return n1 / n2
    except ZeroDivisionError:
        return "0으로 나눌 수 없습니다"


print(proc_divide(10, 2))
print(proc_divide(5, 0))
````

```
[문제2]
매개변수로 지정한 텍스트 파일 경로를 조회하여 읽어오되, 해당 파일이 디렉토리에 없어도 예외 크래시 없이 "파일이 존재하지 않습니다."를 출력하는 proc_safe_open 함수를 구현하세요.

[출력결과]
파일이 존재하지 않습니다.

```

````python
def proc_safe_open(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f: #encoding 옵션에 대해서 좀 알아봐야겠다
            print(f.read())
    except FileNotFoundError:
        print("파일이 존재하지 않습니다.")


proc_safe_open("main2.py")
proc_safe_open("sample.txt")
````

```
[문제5]
리스트 내의 문자열 데이터 숫자들을 모두 더하는 과정에서 문자가 섞여 연산 오류(TypeError)가 발생하면, 무시하고 숫자 형태의 데이터만 합산하여 결과를 반환하는 proc_sum_list 함수를 구현하세요.


[출력결과]
60

```




````python
def proc_sum_list(l):
    total = 0
    for i in l:
        try:
            total += i
        except TypeError as e:
            pass
    return total

k = [10,20,30,'a',False]
proc_sum_list(k)
````

```
[문제8]
하나의 try 블록 안에서 두 가지 이상의 에러가 발생할 여지가 있을 때, ValueError와 ZeroDivisionError를 멀티 except 구문으로 연쇄 캐칭하여 각각 다른 에러 안내 메시지를 출력하는 proc_multi_exception 함수를 구현하세요.

[출력결과]
에러: 숫자로 변환할 수 없는 입력입니다.
에러: 0으로 나눌 수 없습니다.

```

````python
def proc_multi_exception():
    try:
        x = int(input())
        if x==0:
            raise ZeroDivisionError
    except ValueError as e:
        print("숫자로 변환할 수 없는 입력입니다.")
    except ZeroDivisionError as e:
        print("0으로 나눌 수 없습니다.")

proc_multi_exception()
````

```
[문제9]
문자열 형식의 숫자를 받아 나눗셈 연산을 수행하되, 예외가 발생하지 않았을 때만 작동하는 else 블록을 결합하여 "연산에 정상 성공했습니다." 문구와 결과값을 출력하는 proc_success_check 함수를 구현하세요.

[출력결과]
연산에 정상 성공했습니다. 결과: 10.0

```

````python
def proc_success_check(string):
    try:
        n = int(string)/2
    except ValueError as e:
        print(e)
    else:
        print(f"연산에 정상 성공했습니다. 결과 :{n}")

proc_success_check('20')
````

```
[문제10]
try-except 구조로 파일 처리 기능을 수행하면서, 작업 도중 예외 에러가 뜨거나 안 뜨거나 상관없이 언제나 함수 탈출 직전에 실행되어 확실하게 시스템 자원을 반환(close)해 주는 finally 블록을 포함한 proc_final_close 함수를 구현하세요.

[출력결과]
예외 핸들링: 파일을 찾지 못해 안전 모드로 전환합니다.
finally 블록: 파일 스트림 자원 체크를 완수합니다.
```


````python
def proc_final_close():
    try:
        f = open("source/sample.txt", "r")
        print(f.read())
    except FileNotFoundError:
        print("예외 핸들링: 파일을 찾지 못해 안전 모드로 전환합니다.")
    finally:
        print("finallt 블록: 파일 스트림 자원 체크를 완수합니다.")
        if f is not None:
            f.close()


proc_final_close()
````

````python
def proc_file_copy(fn,cn):
    with open(fn,'r') as f_f:
        data = f_f.read()
    with open(cn,'w') as f_c:
        f_c.write(data)

proc_file_copy(복사할 파일, 복사될 파일의 이름)
````
