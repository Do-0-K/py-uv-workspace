### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.

| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |

`````python
class B:
  x = 1

  def xx(self, x): # 인스턴스 메소드
    self.x = x

  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x

  @staticmethod
  def zz(a, b):  # 스테이틱 메소드
    print(a + b)

  @staticmethod  # 스테이틱 메소드
  def kk():
    print(x)
`````

`````python
B.yy(10)
`````

`````python
B.x
`````

`````python
B.zz(3,4)
`````

`````python
B.kk()
`````

`````python
class Book:
    def __init__(self,title=None,author=None,page=None): # private 처럼 사용하려면 self.__변수이름
        self.__title = title
        self.__author = author
        self.__page = page
    
    # setter 함수
    def setTitle(self,title):
        self.__title = title
    def setAuthor(self,author):
        self.__author = author
    def setPage(self,page):
        self.__page = page

    # getter 함수
    def getTitle(self):
        return self.__title
    def getAuthor(self):
        return self.__author
    def getPage(self):
        return self.__page

    def toString(self):
        return f"{self.__title}의 저자는 {self.__author}이고 {self.__page}페이지로 구성됨"
`````

`````python
# 부모클래스
class Parent:
  x = 'parent x'  # 클래스 변수
  y = 'parent y'  # 클래스 변수

  @classmethod
  def output(cls):
    return cls.x

  def test(self):
     return 10

#자식클래스
class Child(Parent):
  x = 'child x' # 클래스 변수

  def display(self):
    print('display')
    print(Parent.output())
    print(Child.output())
    print(Parent.test(self))
    print(Child.test(self))
`````

## overriding(오버라이딩)
부모가 가진 기능을 변경하는 것을 overriding이라고 한다.
double underbar 같은 것들은 overriding해서 객체의 기능을 바꿀 수 없다.

`````python
class Animal:
    def __init__(self,name):
        self.name = name
    def make_sound(self):
        # 자식 클래스들이 반드시 재정의(오버라이딩)하도록 유도하는 부모의 기본 메서드
        return "알 수 없는 동물의 소리입니다"
    
class Dog(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "멍멍!"
    
class Cat(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "야옹!"
`````

`````python
from abc import ABC, abstractmethod

class Shape(ABC): # 추상 메소드를 가지고 있는 클래스이기 때문에 인스턴스가 불가 (객체 생성)

    def __init__(self, width, height):
        self.width = width
        self.height = height

    @abstractmethod # 추상 메소드 데코레이터
    def area(self): # 객체에 맞게 오버라이딩 하도록 추상 메소드로 선언.
      pass          # 해당 클래스를 상속받는 자식 클래스들은 오버라이딩을 안하면 오류가 나도록 강제성 부여

class Rect(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)

    def area(self):
      return self.width * self.height  

class Triangle(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)

    def area(self):
      return self.width * self.height * 0.5
`````

`````python
s = Shape(3,2)
`````

`````python
r = Rect(2, 5)
print(f'사각형 넓이 : {r.area()}')
t = Triangle(2, 5)
print(f'삼각형 넓이 : {t.area()}')
`````
